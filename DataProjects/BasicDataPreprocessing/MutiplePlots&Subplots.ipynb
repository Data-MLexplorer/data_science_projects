```python
tips=sns.load_dataset('tips')
print(tips.head())
```

```python
print(tips.describe())
```

```python
print(tips.info())
```

```python
fig=plt.figure()
axes1=fig.add_subplot(1,1,1)
axes1.hist(data=tips, x='total_bill', bins=244)
axes1.set_title('Histogram of Total Bill')
axes1.set_ylabel('Frequency')
axes1.set_ylabel('Total Bill')
plt.show()
```

```python
scatter_plot=plt.figure()
axes1=scatter_plot.add_subplot(1,1,1)
axes1.scatter(data=tips, x='total_bill', y='tip')
axes1.set_title('Scatter plot of Total Bill vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
plt.show()
```

```python
boxplot=plt.figure()
axes1=boxplot.add_subplot(1,1,1)
axes1.boxplot(
    x=[
        tips.loc[tips["sex"]=="Female", "tip"],
        tips.loc[tips["sex"]=="Male", "tip"]
    ],
    tick_labels=["Female", "Male"]
)
axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
axes1.set_title('Boxplot of Tips by Gender')
plt.show()
```

```python
colors = {
    "Female": "#f1a340",
    "Male": "#998ec3"
}
scatter_plot=plt.figure()
axes1=scatter_plot.add_subplot(1,1,1)
axes1.scatter(
    data=tips,
    x='total_bill',
    y='tip',
    s=tips["size"]**2*10,
    c=tips['sex'].map(colors),
    alpha=0.5
)
axes1.set_title('Colored by sex and sized bu size')
axes1.set_xlabel('Total bill')
axes1.set_ylabel('Tip')
scatter_plot.suptitle("Total Bill vs Tip")
plt.show()
```

```python
tips=sns.load_dataset("tips")
sns.set_context("paper")
print(tips.head())
```

```python
fig, ax = plt.subplots()  # Correct way to get a figure and axis
sns.histplot(data=tips, x='total_bill', ax=ax)  # Pass 'ax' to seaborn
ax.set_title('Total Bill Histogram')

plt.show()
```

```python
# den, ax=plt.subplot()
# sns.kdeplot(data=tips, x="total_bill", ax=ax)
# ax.set_title('Total Bill Density')
# ax.set_xlabel('Total Bill')
# ax.set_ylabel('Unit Probability')
# plt.show()

fig, ax = plt.subplots()  # Correct: creates a figure and an axis
sns.kdeplot(data=tips, x="total_bill", ax=ax)  # Plot the density plot on the axis
ax.set_title('Total Bill Density')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Unit Probability')

plt.show()
```

```python
# rug, ax=plt.subplot()
# sns.histplot(data=tips, x='total_bill', ax=ax, bins=5, alpha=0.6)
# sns.rugplot(data=tips, x='total_bill', ax=ax, height=0.1, linewidth=1)
# ax.set_title('Rug plot gistogram of total bill')
# ax.set_label("Rug Plot", fontsize=12)
# plt.show()

fig, ax = plt.subplots()

# Create histogram with bins and transparency
sns.histplot(data=tips, x='total_bill', ax=ax, bins=5, alpha=0.6)

# Add a rug plot
sns.rugplot(data=tips, x='total_bill', ax=ax, height=0.1, linewidth=1)

# Set title and labels correctly
ax.set_title('Rug Plot Histogram of Total Bill')
ax.set_xlabel("Total Bill", fontsize=12)  # Correct way to set x-axis label
ax.set_ylabel("Frequency", fontsize=12)   # Set y-axis label

# Show the plot
plt.show()
```

```python
fig=sns.displot(data=tips, x="total_bill", kde=True, rug=True)
fig.set_axis_labels(x_var="Total Bill", y_var="Count")
fig.figure.suptitle('Distraction of Total Bill')
plt.show()

```

```python
# count, ax=plt.subplot()
# sns.countplot(data=tips, x='day', palette='viridis', ax=ax)
# ax.set_title('Count of Days')
# ax.set_xlabel('Days of the week')
# ax.set_ylabel('Frequency')
# plt.show()

fig, ax = plt.subplots()

# Create a count plot
sns.countplot(data=tips, x='day', palette='viridis', ax=ax)

# Set title and labels correctly
ax.set_title('Count of Days')
ax.set_xlabel('Days of the Week')
ax.set_ylabel('Frequency')

# Show the plot
plt.show()
```

```python
reg, ax=plt.subplots()
sns.regplot(data=tips, x='total_bill', y='tip', ax=ax)
ax.set_title('Regression plot of Total bill and Tip')
ax.set_xlabel('total bill')
ax.set_ylabel('tip')
plt.show()
```

```python
joint=sns.jointplot(data=tips, x='total_bill', y='tip')
joint.set_axis_labels(xlabel="Total Bill", ylabel='Tip')
joint.figure.suptitle('Joint plot of Total Bill and Tip', y=1.03)
plt.show()
```

```python
hexbins = sns.jointplot(data=tips, x='total_bill', y='tip', kind='hex')
hexbins.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
hexbins.figure.suptitle('Hexbin Plot of Total Bill', y=1.03)
plt.show()
```

```python
ebola=pd.read_csv('country_timeseries.csv')
print(ebola.columns)
```

```python
print(ebola.iloc[:5,[0,1,2,10]])
```

```python
ebola_long=ebola.melt(id_vars=['Date','Day'])
print(ebola_long)
```

```python
variable_split=ebola_long.variable.str.split('_')
print(variable_split[:5])
```

```python
print(type(variable_split[0]))
```

```python
status_values=variable_split.str.get(0)
country_values=variable_split.str.get(1)
print(status_values)
print(country_values)
```

```python
ebola_long['status']=status_values
ebola_long['country']=country_values
print(ebola_long)
```

```python
ebola_long = ebola.melt(id_vars=['Date','Day'])
variable_split=ebola_long.variable.str.split('_',expand=True)
print(variable_split)
```

```python
ebola_long[['status','country']]=variable_split
print(ebola_long)
```

```python
weather=pd.read_csv('weather.csv')
print(weather.iloc[:5,:11])
```

```python
weather_melt=weather.melt(
    id_vars=["id","year","month","element"],
    var_name="day",
    value_name="temp",
)
print(weather_melt)
```

```python
weather_tidy=weather_melt.pivot_table(
    index=['id','year','month','day'],
    columns='element',
    values='temp',
)
print(weather_tidy)
```

```python
weather_tidy_flat=weather_tidy.reset_index()
print(weather_tidy_flat)
```

```python
weather_tidy=(
    weather_melt.pivot_table(
      index=['id','year','month','day'],
      columns= 'element',
      values='temp'
    )
    .reset_index()
)
print(weather_tidy)
```
