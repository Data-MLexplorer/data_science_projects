```python
import seaborn as sns
import matplotlib.pyplot as plt

# Loading the tips data set from the sns library :)
tips = sns.load_dataset("tips")

 
#figsize function takes two param
plt.figure(figsize=(10, 7))
#configuring the figure.
sns.kdeplot(x=tips['total_bill'], y=tips['tip'], fill=True, cmap="Blues")

# Add titles and labels with font size parameters.
plt.title("Kernel Density Plot for Total Bill vs. Tip", fontsize=16)
plt.xlabel("Total Bill", fontsize=14)
plt.ylabel("Tip", fontsize=14)

# Show the plot or figure
plt.show()
```
